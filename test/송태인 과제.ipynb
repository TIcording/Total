```python
class DAO:
    def __init__(self):
        self.db = None
    
    def connect(self):
        self.db = MySQLdb.connect('localhost', 'root', '1234', 'test')
        
    def disconnect(self):
        self.db.close()
    
    def insert(self, it):
        self.connect()
        cur = self.db.cursor()
        sql ='insert into student values(%s,%s,%s,%s,%s,now())'
        data = int((it.getnumber(), it.getname(), it.gethp(), it.getemail(),it.address()))
        print(data)
        cur.execute(sql, data)
        self.db.commit()
        self.disconnect()     
        
    def selectAi(self):
        self.connect()
        cur = self.db.cursor()
        sql = 'select name, age, gender from ai'
        cur.execute(sql)
        row = cur.fetchall()
        self.disconnect()
        return row


    def searchAi(self, name):
        self.connect()
        cur = self.db.cursor()
        sql = 'select name, age, gender from ai where name=%s'
        data = (name,)     
        cur.execute(sql, data)
        row = cur.fetchone()
        self.disconnect()
        return row

    
    def updateAi(self, it):
        self.connect()
        cur = self.db.cursor()
        sql = 'update ai set age=%s, gender=%s where name=%s'
        data = (it.getAge(), it.getGender(), it.getName())
        result = cur.execute(sql, data)
        self.db.commit()
        if result > 0:
            print('수정되었습니다')
        else:
            print('해당 학생이 없습니다')
        self.disconnect()

    
    def deleteAi(self, name):
        self.connect()
        cur = self.db.cursor()
        sql = 'delete from ai where name=%s'
        data = (name,)
        result = cur.execute(sql, data)
        self.db.commit()
        if result > 0:
            print('삭제되었습니다')
        else:
            print('해당 학생이 없습니다')
        self.disconnect()
```
